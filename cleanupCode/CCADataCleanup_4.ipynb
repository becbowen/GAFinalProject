```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
ccas = pd.read_excel("../rawData/CCAs_2006_2019.xlsx")
ccas.head(3)
```

```python
ccas = ccas[["Code", "Category1", "Textbox14"]]
```

```python
ccas.head()
```

```python
ccas["Year"] = ccas["Code"].str[-4:]
ccas.head()
```

```python
ccas["Year"] = ccas["Year"].apply(np.int64)
```

```python
ccas["Site"] = ccas["Code"].str[0:-27]
ccas.tail()
```

```python
ccas = ccas[["Site","Year","Category1","Textbox14"]]
```

```python
ccas.head()
```

```python
ccas = ccas.groupby(by=['Site', 'Year', "Category1"]).sum().reset_index()
```

```python
ccas.head(3)
```

```python
ccas = ccas[(ccas.Category1 != "SCWE")].reset_index()
ccas.head(3)
```

```python
ccasP = ccas[(ccas["Category1"] == "PI&R")].reset_index()
ccasP.head()
```

```python
ccas1 = ccasP[["Textbox14"]]
ccas1["PIR"] = ccas1["Textbox14"]
ccas1 = ccas1[["PIR"]]
ccas1.head()
```

```python
ccasH = ccas[(ccas["Category1"] == "Human Performance")].reset_index()
ccasH.head()
```

```python
ccas2 = ccasH[["Textbox14"]]
ccas2["HU"] = ccas2["Textbox14"]
ccas2 = ccas2[["HU"]]
ccas2.head()
```

```python
ccas3 = pd.concat([ccas2, ccas1], axis=1)
ccas3.head()
```

```python
ccas3.shape
```

```python
ccas.shape
```

```python
ccas = pd.concat([ccasH,ccas3], axis = 1)
ccas.head()
```

```python
ccas = ccas[["Site","Year","HU","PIR"]]
ccas.head()
```

```python
ccas.shape
```

```python
ccas[ccas["Site"] == "Oyster Creek"]
```

```python
ccas.drop([190,191,192,193,194,195,414,415,416,417,418,419,722,723,724,725,726,727,849,850,851,852,853,319,320,321,545], inplace = True)
```

```python
ccas.shape
```

```python
ccas["Site"].unique()
```

```python
ccas["Site"] = ccas["Site"].map({
        'D.C. Cook': 'Cook',
        'Davis-Besse': 'Davis Besse',
        'Columbia Generating Station': 'Columbia'
        }).fillna(ccas["Site"])
```

```python
ccas = ccas.sort_values(by =['Site', 'Year']).reset_index()
ccas = ccas[["Site","Year","HU","PIR"]]
```

```python
ccas.tail(3)
```

```python
ccas.to_excel("ccaData.xlsx")
```
