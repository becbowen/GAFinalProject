```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
ssff = pd.read_excel("LER_SSFFs.xlsx")
```

```python
ssff.head()
```

```python
ssff = ssff[["Plant Name", "Event Date", "Accession #"]]
ssff.head()
```

```python
ssff["Event Date"].dtype
```

```python
ssff["Year"] = ssff["Event Date"].str[-4:]
ssff.head()
```

```python
ssff["Year"] = ssff["Year"].apply(np.int64)
ssff["Year"].dtype
```

```python
ssff["Plant Name"].unique()
```

```python
#plants to sites conversion only has one plant name, so need plant name to only be the first name listed in the column
ssff["Plant Name"] = ssff["Plant Name"].str.split(",").str[0]
ssff.tail()
```

```python
convert = pd.read_excel("plantsToSitesLER.xlsx")
```

```python
#make this a dictionary 
site_dict = convert.set_index('Plant')['Site'].to_dict()
```

```python
#apply the unit-site conversion dictionary site_dict to the am_dataframe to get a column by site
ssff['Site']= ssff['Plant Name'].map(site_dict)
```

```python
ssff[ssff["Site"].isnull()]
```

```python
ssff.tail()
```

```python
ssff = ssff[["Site", "Year", "Accession #"]]
ssff.head()
```

```python
ssff = ssff.groupby(by=['Site', 'Year']).count().reset_index()
```

```python
ssff.head()
```

```python
ssff.shape
```

```python
ssff["SSFF"] = ssff["Accession #"]
ssff = ssff[["Site", "Year", "SSFF"]]
ssff.head()
```

```python
data = pd.read_excel("data.xlsx", index_col = 0)
data.head()
```

```python
new_ssff = data[["SiteName","Year"]]
new_ssff["Year"] = new_ssff["Year"].apply(np.str)

```

```python
new_ssff["SiteYear"] = new_ssff["SiteName"] + new_ssff["Year"]
```

```python
new_ssff.head()
```

```python
ssff["Year"] = ssff["Year"].apply(np.str)
ssff["SiteYear"] = ssff["Site"] + ssff["Year"]
ssff.head()
```

```python
ssff_merge = pd.merge(new_ssff, ssff, how = "left", left_on="SiteYear", right_on="SiteYear")
ssff_merge.head()
```

```python
ssff_merge["SSFF"] = ssff_merge["SSFF"].fillna(0)
```

```python
ssff_merge.head()
```

```python
ssff = ssff_merge[["SiteName","Year_x","SSFF"]]
ssff.head()
```

```python
ssff["Year"] = ssff["Year_x"]
ssff = ssff[["SiteName", "Year", "SSFF"]]
```

```python
ssff.head()
```

```python
ssff["Year"] = ssff["Year"].apply(np.int64)
ssff["SSFF"] = ssff["SSFF"].apply(np.int64)
ssff.head()
```

```python
ssff = ssff.sort_values(by =['SiteName', 'Year']).reset_index()
```

```python
ssff.shape
```

```python
ssff = ssff[["SiteName","Year","SSFF"]]
```

```python
ssff.tail()
```
