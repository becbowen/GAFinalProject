```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
allCCAs = pd.read_excel("CCAs_2006_2019.xlsx")
allCCAs.head(2)
```

```python
allCCAs = allCCAs[["Code","Category1","Title","MidCycleTotal","Textbox14"]]
allCCAs.head(2)
```

```python
allCCAs.rename(columns = {'Code':'Site',"Category1":"Area","Title":"Aspect","Textbox14":"EOCTotal"}, inplace = True)
allCCAs.head(2)
```

```python
allCCAs["Year"] = allCCAs["Site"].str[-4:]
allCCAs.tail(2)
```

```python
allCCAs["Year"].astype(np.int64)
```

```python
allCCAs.Site = pd.Index(map(lambda x : str(x)[:-27], allCCAs.Site))
allCCAs.tail(2)
```

```python
allCCAs.head(2)
```

```python
allCCAs['Aspect'] = [x.split(" ")[0] for x in allCCAs['Aspect']]
allCCAs.tail(2)
```

```python
allCCAs.head(2)
```

```python
convert = pd.read_excel("CCACrossWalk.xlsx")
site_dict = convert.set_index('OldAspect')['NewAspect'].to_dict()
allCCAs['NewAspect']= allCCAs['Aspect'].map(site_dict)
```

```python
allCCAs.head()
```

```python
allCCAs.tail()
```

```python
allCCAs["NewAspect"].fillna(allCCAs["Aspect"], inplace = True)
```

```python
allCCAs.tail()
```

```python
allCCAs = allCCAs[["Site","Year","Area","NewAspect","MidCycleTotal","EOCTotal"]]
allCCAs.head()
```

```python
newAspects = allCCAs["NewAspect"].unique()
newAspects
```

```python
CCAAspects = allCCAs.groupby(by=['Site', 'Year', "NewAspect"]).sum().reset_index()
CCAAspects.head(23)
```
